#Travel Planning Agent

A example of an hierachical Agent application consisting of multiple agents which are called in a sequence to plan trvel to a "Destination" city from an "Starting" city.

Agents,

-- Air travel agent that finds the best 3 airline options

-- Hotel agent that identifies the Top 3 hotels

-- Concierge agent that identifies leisure activities

-- Planner agent that prepares a datewise formatted report with links for booking

```python
from crewai import Agent, Task, Crew, Process
from crewai_tools import SerperDevTool, WebsiteSearchTool
from IPython.display import Markdown, display
```

# Inputs section


**Please set the required Inputs**

```python
# Inputs - Please provide the inputs
starting_city = "San Francisco"
destination_city = "Los Angeles"
start_date = "12/20/2025"
end_date = "12/31/2025"
```

# Define Agents to Plan the Travel

```python
# Create the Air Travel Agent
travel_agent = Agent(
    name="Travel",
    role="Air Travel Booking Agent",
    goal="Get the best possible Air ticket fares form '{starting_city}' to '{destination_city}' for travel on starting '{start_date}' return on '{end_date}'",
    backstory="An experienced air travel agent skilled in finding the best possible air fares",
    tools=[SerperDevTool(), WebsiteSearchTool()],
    #verbose=True
)

# Create the Hotel Booking Agent
hotel_agent = Agent(
    name="Hotel",
    role="Hotel Booking Agent",
    goal="Get the best possible Hotel availibility in '{destination_city}' from {start_date}' to '{end_date}'",
    backstory="An experienced travel consultant who can research & get a good deal on a Hotel bookings",
    tools=[SerperDevTool(), WebsiteSearchTool()],
   #verbose=True
)

# Create the Concierge Agent
concierge_agent = Agent(
    name="Concierge",
    role="Leisure Activities Planner",
    goal="Indentify & plan for leisure activities including tours, dinner, theater in '{destination_city}' from {start_date}'+1 to '{end_date}'-1",
    backstory="An organized planner who crafts a exciting tourism activities in the given place which gives exposure to the local culture, surroundings",
    #verbose=True
)

# Create the Travel Planner Agent
planner_agent = Agent(
    name="Planner",
    role="Activities planner",
    goal="Summarize research findings clearly, in an organized and easy to digest way",
    backstory="An organized planner who crafts well-structured plan inculding Air tickets, Hotel bookings & a daily acitvities plan",
    #verbose=True
)

```

# Define Tasks for the Agents

```python
# Define tasks
research_task_travel = Task(
    description=f"Get the best possible Air ticket fares from '{starting_city}' to '{destination_city}' for travel on starting '{start_date}' return on '{end_date}'.",
    expected_output=f"List the Top 3 air ticket options for travel from '{starting_city}' to '{destination_city}' for travel on starting '{start_date}' return on '{end_date}'.Provide links to do the flight bookings. List each option on seperate lines",
    agent=travel_agent
)

research_task_hotel = Task(
    description=f"Get the best possible Hotel rates for '{destination_city}' for travel on starting '{start_date}' return on '{end_date}'.",
    expected_output=f"List the Top 3 Hotel rates for '{destination_city}' for travel on starting '{start_date}' return on '{end_date}'. Provide links to do the hotel bookings. List each option on seperate lines",
    agent=hotel_agent
)

research_task_concierge = Task(
    description=f"Identify & plan for leisure activities including tours, dinner, theater in '{destination_city}' from {start_date}'+1 to '{end_date}'-1. ",
    expected_output=f"Bullet points activities identified alongwith the costs. Provide links to do bookings.",
    agent=concierge_agent
)

write_task = Task(
    description=f"Write a structured report on Travel plan to '{destination_city}' from '{starting_city}', using research findings on Air tickets, Hotel Booking & Leisure activities wtih costs.",
    expected_output=f"A markdown summary of the Travel plan to '{destination_city}' with clear sections, for each day. Include any travel requirements & hints. Add a Budgetory statement, including the above top Airfare, top Hotel & Activities expenses & Grand Total, each on seperate lines. At the end, please include an inspirational section on why the {destination_city} is an ideal place for a vacation.",
    agent=planner_agent,
    context=[research_task_travel, research_task_hotel, research_task_concierge]
)

```

# Assemble the crew and run tasks


```python
crew = Crew(
    agents=[travel_agent, hotel_agent, concierge_agent, planner_agent],
    tasks=[research_task_travel, research_task_hotel, research_task_concierge, write_task],
    process=Process.sequential
)

result = crew.kickoff()

# Display the final report
display(Markdown(result.raw))
```
